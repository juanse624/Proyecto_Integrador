# ➡️**Selección del Modelo Final**

Para la selección de la mejor franja horaria para contactar clientes morosos, la validación del modelo y la optimización de hiperparámetros son pasos cruciales para garantizar la efectividad y confiabilidad de la solución.

¿Por qué es necesaria la validación?

- **Evitar el sobreentrenamiento**: Un modelo sobreajustado se adapta perfectamente a los datos de entrenamiento, pero no generaliza bien a nuevos datos. La validación nos permite identificar este problema y ajustar el modelo para que sea más robusto y generalizable.

- **Seleccionar el mejor modelo**: Al entrenar diferentes modelos con distintos algoritmos o configuraciones, la validación nos ayuda a comparar su rendimiento y elegir el que mejor se desempeña en los datos de validación.

- **Evaluar la precisión y confiabilidad**: La validación nos proporciona métricas de desempeño, como la precisión, la sensibilidad y la especificidad, que nos permiten evaluar la capacidad del modelo para identificar correctamente las mejores franjas horarias de cobro.

- **Ajustar los hiperparámetros**: Los hiperparámetros controlan el comportamiento del modelo, como la tasa de aprendizaje o el número de árboles en un bosque aleatorio. La optimización de hiperparámetros mediante técnicas como la búsqueda en rejilla o la búsqueda aleatoria nos permite encontrar la configuración que maximiza el rendimiento del modelo.

Al trabajar con modelos como Random Forest y Decision Tree, la validación y la optimización de hiperparámetros son particularmente importantes debido a su flexibilidad y capacidad de adaptarse a datos complejos.

- `Random Forest`: La cantidad de árboles en el bosque, la profundidad máxima de los árboles y las características aleatorias por división son algunos de los hiperparámetros clave a considerar.
- `Decision Tree`: La profundidad máxima del árbol, el criterio de división y las opciones de poda son aspectos importantes a optimizar.

La validación del modelo y la optimización de hiperparámetros son procesos esenciales para garantizar la efectividad y confiabilidad de un modelo de Machine Learning para la selección de la mejor franja horaria de cobranza. Al evitar el sobreajuste, seleccionar los hiperparámetros óptimos, mejorar la confiabilidad y la interpretabilidad del modelo, se obtienen decisiones más informadas y efectivas para la recuperación de cartera, optimizando así los procesos de cobranza de la empresa.

**Estrategias de Ensamble**

La aplicación del estrategias de ensamble en el modelo de Machine Learning para la selección de franjas horarias puede conducir a una mayor precisión, robustez, generalización y reducción de errores.

A continuación, se detallan algunas razones clave por las que su aplicación sería beneficiosa:

1. **Mayor precisión y robustez**: Las estrategias de ensamble combinan las predicciones de múltiples modelos, lo que reduce el riesgo de sesgos individuales y aumenta la precisión general. Al considerar diversos enfoques y perspectivas, el modelo final es menos susceptible a errores o inconsistencias en un solo modelo.

2. **Mejor manejo de la incertidumbre**: Los modelos de aprendizaje automático a menudo se enfrentan a la incertidumbre inherente en los datos. Aplicar una estrategia de ensamble ayuda a mitigar esta incertidumbre al promediar o combinar las predicciones de diferentes modelos. Esto puede conducir a resultados más confiables y consistentes, especialmente en situaciones con datos limitados o ruidosos.

3. **Aprovechamiento de las fortalezas de diferentes algoritmos**: Los distintos algoritmos de aprendizaje automático tienen diferentes fortalezas y debilidades. El aprendizaje en conjunto permite combinar las fortalezas de varios modelos para crear un modelo más robusto y versátil. Por ejemplo, un modelo Random Forest puede ser bueno para identificar patrones complejos, mientras que un Decision Tree puede ser más eficiente en la interpretación de las reglas de decisión.

4. **Mejora de la generalización**: Un problema común en el Machine Learning es el sobreentrenamiento, donde el modelo se adapta demasiado a los datos de entrenamiento y no funciona bien en nuevos datos. El aprendizaje en conjunto puede ayudar a prevenir el sobreentrenamiento al promover la diversidad entre los modelos y evitar que un solo modelo domine las predicciones.

5. **Reducción del error de tipo II**: El error de tipo II ocurre cuando un modelo no identifica correctamente un caso positivo (por ejemplo, un cliente moroso que no se llama en la mejor franja horaria). El aprendizaje en conjunto puede ayudar a reducir este tipo de error al aumentar la sensibilidad del modelo para detectar casos positivos.

# ➡️ **Introduccion**

**Ensamble Homogéneo**

El ensamble homogéneo utiliza múltiples instancias del mismo tipo de modelo para mejorar el rendimiento general. Algunas técnicas comunes incluyen:

- `Bagging (Bootstrap Aggregating)`: Combina las predicciones de varios clasificadores entrenados en diferentes subconjuntos del mismo dataset.
Ejemplo: Random Forest, donde múltiples árboles de decisión son entrenados en diferentes subconjuntos de datos con reemplazo.
- `Boosting`: Combina predicciones de varios modelos débiles de manera secuencial, donde cada modelo intenta corregir los errores de su predecesor.
Ejemplo: `AdaBoost`, `Gradient Boosting`.
- `Stacking (Stacked Generalization)`: Usa múltiples modelos de nivel base (del mismo tipo) y un modelo meta para combinar sus predicciones.
Ejemplo: Varias instancias de SVM combinadas con un modelo de regresión logística como meta-modelo.

**Ensamble Heterogéneo**

El ensamble heterogéneo combina diferentes tipos de modelos para aprovechar sus diversas fortalezas y debilidades.

- `Bagging Heterogéneo`: Similar a bagging, pero utiliza diferentes tipos de clasificadores.
Ejemplo: Combinar `Random Forest`, `SVM` y `KNN` en un solo modelo de ensamble.
- `Boosting Heterogéneo`: Implementa boosting pero con diferentes tipos de modelos en cada iteración.
Ejemplo: Una variación personalizada de boosting que incluye árboles de decisión y regresión logística.
- `Stacking Heterogéneo` : Usa diferentes modelos de nivel base y un modelo meta para combinar sus predicciones.
Ejemplo: Combinación de `Random Forest`, `SVM`, y `KNN`, con un `meta-modelo` de regresión logística o MLP.

Aplicar estrategias de ensamble es importante en problemas de clasificación como nuestro caso,  ya que podemos tener los siguientes beneficios:

- **Mejora del Rendimiento**: Los modelos de ensamble suelen superar el rendimiento de los modelos individuales al reducir la varianza (bagging) o el sesgo (boosting).
- **Robustez**: Los modelos de ensamble son menos propensos a sobreajustarse a los datos de entrenamiento, ya que combinan múltiples predicciones.
- **Estabilidad**: Proveen una mayor estabilidad y confiabilidad en las predicciones, ya que los errores de los modelos individuales tienden a cancelarse entre sí.
- **Mitigación de Debilidades**: Los enfoques heterogéneos pueden combinar modelos que tienen diferentes fortalezas y debilidades, logrando un modelo final más equilibrado y robusto.

## Random Forest

`Random Forest` es un algoritmo de aprendizaje automático supervisado que se utiliza comúnmente para tareas de clasificación y regresión. Es un método de conjunto que combina múltiples árboles de decisión para crear un modelo predictivo más robusto y preciso.

**Funcionamiento:**

**Creación de `árboles de decisión`:** Se entrenan varios árboles de decisión de forma independiente utilizando subconjuntos aleatorios de los datos de entrenamiento.

**Selección aleatoria de características:** En cada nodo del árbol, se considera un subconjunto aleatorio de características para determinar la mejor división.

**Voto mayoritario:** Para realizar una predicción, cada `árbol de decisión` del bosque emite un voto, y la clase con la mayoría de votos se asigna a la nueva instancia.

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

## Decision Tree

`Decision Tree` es un algoritmo de aprendizaje automático supervisado comúnmente utilizado para tareas de clasificación y regresión. Son una forma intuitiva y fácil de entender de representar el proceso de toma de decisiones y se pueden usar para resolver una amplia gama de problemas.

**Funcionamiento:**

**Selección de la característica raíz:** Se elige la característica más importante del conjunto de datos para dividir el nodo raíz del árbol.

**Creación de ramas:** Para cada valor posible de la característica raíz, se crean ramas que representan subconjuntos de los datos.

**Repetición:** El proceso se repite en cada nodo interno del árbol, seleccionando la mejor característica y creando ramas para sus valores, hasta que se alcanza un criterio de parada (por ejemplo, profundidad máxima del árbol o pureza suficiente del nodo).

**Predicción:** Para clasificar una nueva instancia, se recorre el árbol desde la raíz hasta una hoja siguiendo los valores de sus características. La clase de la hoja se asigna como la predicción para la instancia.

```python
param_grid_dt = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
dt = DecisionTreeClassifier()
grid_search_dt = GridSearchCV(dt, param_grid_dt, cv=2)
grid_search_dt.fit(X_train, y_train)
```

# ➡️ **Utilizar algoritmos que apliquen tanto estrategias de ensamble homogéneas como heterogéneas.**

## Ensamble Homogéneo (Bagging)

`Bagging` es una técnica de ensamble homogéneo que se utiliza para mejorar la estabilidad y la precisión de los modelos de aprendizaje automático. Funciona entrenando múltiples modelos base independientes, generalmente del mismo tipo (por ejemplo, árboles de decisión), en diferentes subconjuntos aleatorios del conjunto de datos de entrenamiento. Luego, combina las predicciones de estos modelos base para obtener una predicción final.

**Funcionamiento:**

Selecciona aleatoriamente múltiples muestras (con reemplazo) del conjunto de entrenamiento.

- Entrena un modelo base en cada una de estas muestras.

- Combina las predicciones de todos los modelos base mediante votación (en el caso de clasificación) o promedio (en el caso de regresión) para obtener la predicción final.

**Ventajas:**

- Reduce la varianza al promediar las predicciones de múltiples modelos.
Reduce el sobreajuste al utilizar muestras aleatorias y modelos independientes.

**Ejemplo de uso:**

- **Clasificación:** `BaggingClassifier` en sklearn.ensemble para clasificación.

- **Regresión:** `BaggingRegressor` en sklearn.ensemble para regresión.

```python
bagging_rf = BaggingClassifier(base_estimator=RandomForestClassifier(**grid_search_rf.best_params_), n_estimators=10)
bagging_dt = BaggingClassifier(base_estimator=DecisionTreeClassifier(**grid_search_dt.best_params_), n_estimators=10)
adaboost_dt = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(**grid_search_dt.best_params_), n_estimators=50, learning_rate=1.0)
```

## Stacking (Stacked Generalization)

`Stacking` es una técnica de ensamble heterogéneo que combina múltiples modelos base de diferentes tipos para mejorar la predicción. Funciona entrenando varios modelos base independientes y luego utilizando otro modelo (conocido como meta-modelo) para combinar las predicciones de estos modelos base.

Funcionamiento:

- Entrena varios modelos base utilizando diferentes algoritmos de aprendizaje automático.

- Utiliza las predicciones de estos modelos base como características para entrenar un meta-modelo.

- El meta-modelo aprende cómo combinar las predicciones de los modelos base para producir una predicción final.

**Ventajas:**

Puede capturar relaciones complejas entre las características y las etiquetas al utilizar diferentes algoritmos de aprendizaje automático.
Puede ofrecer un mejor rendimiento que los modelos individuales.

**Ejemplo de uso:**

- **Clasificación:** `StackingClassifier` en sklearn.ensemble para clasificación.

- **Regresión:** `StackingRegressor` en sklearn.ensemble para regresión.

```python
estimators = [('rf', RandomForestClassifier(**grid_search_rf.best_params_)),
              ('dt', DecisionTreeClassifier(**grid_search_dt.best_params_))]
stacking = StackingClassifier(estimators=estimators, final_estimator=RandomForestClassifier())
```

```python
results_ensamble = {}

# Itera sobre los modelos de ensamble y calcula las métricas
for name, model in models_ensamble.items():
    start_time = time.time()  # Tomar el tiempo de inicio
    model.fit(X_train, y_train)  # Entrenar el modelo
    end_time = time.time()  # Tomar el tiempo de finalización
    train_time = end_time - start_time  # Calcular el tiempo de entrenamiento

    # Realizar predicciones en el conjunto de prueba
    y_pred = model.predict(X_test)

    # Calcular las métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Almacenar los resultados en el diccionario
    results_ensamble[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1,
        "Confusion Matrix": cm,
        "Training Time (s)": train_time
    }

# Imprimir los resultados de los modelos de ensamble
for name, metrics in results_ensamble.items():
    print(f"Resultados para {name}:")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")
    print("\n")
```

```python
# Definir los rangos de valores para el hiperparámetro que quieres validar
param_range = [10, 50, 100, 200]

# Calcular la curva de validación para el hiperparámetro (por ejemplo, n_estimators en RandomForest)
train_scores, test_scores = validation_curve(mejor_modelo_rf, X_train, y_train, param_name="n_estimators", param_range=param_range, cv=5)

# Calcular los puntajes medios y las desviaciones estándar para los conjuntos de entrenamiento y prueba
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Trazar la curva de validación
plt.figure(figsize=(10, 6))
plt.title("Curva de Validación")
plt.xlabel("Número de Estimadores")
plt.ylabel("Puntuación")
plt.ylim(0.0, 1.1)
lw = 2
plt.plot(param_range, train_scores_mean, label="Entrenamiento", color="darkorange", lw=lw)
plt.fill_between(param_range, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.2, color="darkorange", lw=lw)
plt.plot(param_range, test_scores_mean, label="Validación Cruzada", color="navy", lw=lw)
plt.fill_between(param_range, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.2, color="navy", lw=lw)
plt.legend(loc="best")
plt.show()
```

Eje X: Número de iteraciones de entrenamiento

Eje Y: Score del modelo (`Accuracy`, `Precision`, `Recall`, `F1-score`)

Esto indica que el modelo está aprendiendo efectivamente de los datos de entrenamiento.

- La curva alcanza un pico y luego comienza a disminuir ligeramente, lo que sugiere que el modelo puede estar comenzando a sobreajustarse a los datos de entrenamiento.

- Se observa un ligero `overfitting` en la curva de validación. El modelo está aprendiendo demasiado de los datos de entrenamiento y no está generalizando bien a los datos de validación.

- La curva de validación del modelo `Random Forest` indica un proceso de entrenamiento exitoso con un ligero `overfitting`.
