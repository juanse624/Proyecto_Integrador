- **Ajustar los hiperparmetros**: Los hiperparmetros controlan el comportamiento del modelo, como la tasa de aprendizaje o el nmero de rboles en un bosque aleatorio. La optimizacin de hiperparmetros mediante tcnicas como la bsqueda en rejilla o la bsqueda aleatoria nos permite encontrar la configuracin que maximiza el rendimiento del modelo.

Al trabajar con modelos como Random Forest y Decision Tree, la validacin y la optimizacin de hiperparmetros son particularmente importantes debido a su flexibilidad y capacidad de adaptarse a datos complejos.

La aplicacin del estrategias de ensamble en el modelo de Machine Learning para la seleccin de franjas horarias puede conducir a una mayor precisin, robustez, generalizacin y reduccin de errores.

A continuacin, se detallan algunas razones clave por las que su aplicacin sera beneficiosa:

#  **Introduccion**

**Ensamble Homogneo**

El ensamble homogneo utiliza mltiples instancias del mismo tipo de modelo para mejorar el rendimiento general. Algunas tcnicas comunes incluyen:

- `Bagging (Bootstrap Aggregating)`: Combina las predicciones de varios clasificadores entrenados en diferentes subconjuntos del mismo dataset.
Ejemplo: Random Forest, donde mltiples rboles de decisin son entrenados en diferentes subconjuntos de datos con reemplazo.
- `Boosting`: Combina predicciones de varios modelos dbiles de manera secuencial, donde cada modelo intenta corregir los errores de su predecesor.
Ejemplo: `AdaBoost`, `Gradient Boosting`.
- `Stacking (Stacked Generalization)`: Usa mltiples modelos de nivel base (del mismo tipo) y un modelo meta para combinar sus predicciones.
Ejemplo: Varias instancias de SVM combinadas con un modelo de regresin logstica como meta-modelo.

**Ensamble Heterogneo**

El ensamble heterogneo combina diferentes tipos de modelos para aprovechar sus diversas fortalezas y debilidades.

- `Bagging Heterogneo`: Similar a bagging, pero utiliza diferentes tipos de clasificadores.
Ejemplo: Combinar `Random Forest`, `SVM` y `KNN` en un solo modelo de ensamble.
- `Boosting Heterogneo`: Implementa boosting pero con diferentes tipos de modelos en cada iteracin.
Ejemplo: Una variacin personalizada de boosting que incluye rboles de decisin y regresin logstica.
- `Stacking Heterogneo` : Usa diferentes modelos de nivel base y un modelo meta para combinar sus predicciones.
Ejemplo: Combinacin de `Random Forest`, `SVM`, y `KNN`, con un `meta-modelo` de regresin logstica o MLP.

Aplicar estrategias de ensamble es importante en problemas de clasificacin como nuestro caso,  ya que podemos tener los siguientes beneficios:

- **Mejora del Rendimiento**: Los modelos de ensamble suelen superar el rendimiento de los modelos individuales al reducir la varianza (bagging) o el sesgo (boosting).
- **Robustez**: Los modelos de ensamble son menos propensos a sobreajustarse a los datos de entrenamiento, ya que combinan mltiples predicciones.
- **Estabilidad**: Proveen una mayor estabilidad y confiabilidad en las predicciones, ya que los errores de los modelos individuales tienden a cancelarse entre s.
- **Mitigacin de Debilidades**: Los enfoques heterogneos pueden combinar modelos que tienen diferentes fortalezas y debilidades, logrando un modelo final ms equilibrado y robusto.

## Random Forest

`Random Forest` es un algoritmo de aprendizaje automtico supervisado que se utiliza comnmente para tareas de clasificacin y regresin. Es un mtodo de conjunto que combina mltiples rboles de decisin para crear un modelo predictivo ms robusto y preciso.

**Voto mayoritario:** Para realizar una prediccin, cada `rbol de decisin` del bosque emite un voto, y la clase con la mayora de votos se asigna a la nueva instancia.

## Decision Tree

`Decision Tree` es un algoritmo de aprendizaje automtico supervisado comnmente utilizado para tareas de clasificacin y regresin. Son una forma intuitiva y fcil de entender de representar el proceso de toma de decisiones y se pueden usar para resolver una amplia gama de problemas.

**Funcionamiento:**

**Seleccin de la caracterstica raz:** Se elige la caracterstica ms importante del conjunto de datos para dividir el nodo raz del rbol.

**Creacin de ramas:** Para cada valor posible de la caracterstica raz, se crean ramas que representan subconjuntos de los datos.

**Repeticin:** El proceso se repite en cada nodo interno del rbol, seleccionando la mejor caracterstica y creando ramas para sus valores, hasta que se alcanza un criterio de parada (por ejemplo, profundidad mxima del rbol o pureza suficiente del nodo).

**Prediccin:** Para clasificar una nueva instancia, se recorre el rbol desde la raz hasta una hoja siguiendo los valores de sus caractersticas. La clase de la hoja se asigna como la prediccin para la instancia.

## Ensamble Homogneo (Bagging)

`Bagging` es una tcnica de ensamble homogneo que se utiliza para mejorar la estabilidad y la precisin de los modelos de aprendizaje automtico. Funciona entrenando mltiples modelos base independientes, generalmente del mismo tipo (por ejemplo, rboles de decisin), en diferentes subconjuntos aleatorios del conjunto de datos de entrenamiento. Luego, combina las predicciones de estos modelos base para obtener una prediccin final.

**Funcionamiento:**

Selecciona aleatoriamente mltiples muestras (con reemplazo) del conjunto de entrenamiento.

- Entrena un modelo base en cada una de estas muestras.

- Combina las predicciones de todos los modelos base mediante votacin (en el caso de clasificacin) o promedio (en el caso de regresin) para obtener la prediccin final.

**Ventajas:**

- Reduce la varianza al promediar las predicciones de mltiples modelos.
Reduce el sobreajuste al utilizar muestras aleatorias y modelos independientes.

**Ejemplo de uso:**

- **Clasificacin:** `BaggingClassifier` en sklearn.ensemble para clasificacin.

- **Regresin:** `BaggingRegressor` en sklearn.ensemble para regresin.

```python
bagging_rf = BaggingClassifier(base_estimator=RandomForestClassifier(**grid_search_rf.best_params_), n_estimators=10)
bagging_dt = BaggingClassifier(base_estimator=DecisionTreeClassifier(**grid_search_dt.best_params_), n_estimators=10)
adaboost_dt = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(**grid_search_dt.best_params_), n_estimators=50, learning_rate=1.0)
```

## Stacking (Stacked Generalization)

`Stacking` es una tcnica de ensamble heterogneo que combina mltiples modelos base de diferentes tipos para mejorar la prediccin. Funciona entrenando varios modelos base independientes y luego utilizando otro modelo (conocido como meta-modelo) para combinar las predicciones de estos modelos base.

Funcionamiento:

- Entrena varios modelos base utilizando diferentes algoritmos de aprendizaje automtico.

- Utiliza las predicciones de estos modelos base como caractersticas para entrenar un meta-modelo.

- El meta-modelo aprende cmo combinar las predicciones de los modelos base para producir una prediccin final.

**Ventajas:**

Puede capturar relaciones complejas entre las caractersticas y las etiquetas al utilizar diferentes algoritmos de aprendizaje automtico.
Puede ofrecer un mejor rendimiento que los modelos individuales.

**Ejemplo de uso:**

- **Clasificacin:** `StackingClassifier` en sklearn.ensemble para clasificacin.

- **Regresin:** `StackingRegressor` en sklearn.ensemble para regresin.

```python
estimators = [('rf', RandomForestClassifier(**grid_search_rf.best_params_)),
              ('dt', DecisionTreeClassifier(**grid_search_dt.best_params_))]
stacking = StackingClassifier(estimators=estimators, final_estimator=RandomForestClassifier())
```

```python
results_ensamble = {}

# Itera sobre los modelos de ensamble y calcula las mtricas
for name, model in models_ensamble.items():
    start_time = time.time()  # Tomar el tiempo de inicio
    model.fit(X_train, y_train)  # Entrenar el modelo
    end_time = time.time()  # Tomar el tiempo de finalizacin
    train_time = end_time - start_time  # Calcular el tiempo de entrenamiento

    # Realizar predicciones en el conjunto de prueba
    y_pred = model.predict(X_test)

    # Calcular las mtricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    # Almacenar los resultados en el diccionario
    results_ensamble[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1,
        "Confusion Matrix": cm,
        "Training Time (s)": train_time
    }

# Imprimir los resultados de los modelos de ensamble
for name, metrics in results_ensamble.items():
    print(f"Resultados para {name}:")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")
    print("\n")
```

- La curva alcanza un pico y luego comienza a disminuir ligeramente, lo que sugiere que el modelo puede estar comenzando a sobreajustarse a los datos de entrenamiento.

- Se observa un ligero `overfitting` en la curva de validacin. El modelo est aprendiendo demasiado de los datos de entrenamiento y no est generalizando bien a los datos de validacin.
